Preprocess data

```python
# Mengubah kolom Date menjadi datetime
dfAMD['Date'] = pd.to_datetime(dfAMD['Date'])
dfApple['Date'] = pd.to_datetime(dfApple['Date'])
```

```python
# Mengatur kolom Date sebagai index
dfAMD.set_index('Date', inplace=True)
dfApple.set_index('Date', inplace=True)
```

```python
# Menggunakan hanya kolom Close
dataAMD = dfAMD['Close']
dataApple = dfApple['Close']
```

```python
plt.figure(figsize=(10, 4))
plt.plot(dataAMD, label='Close Price')
plt.title('AMD Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(dataApple, label='Close Price')
plt.title('Apple Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
dataAMD = dataAMD.values.reshape(-1, 1)
```

```python
dataApple = dataApple.values.reshape(-1, 1)
```

```python
# Membuat fungsi untuk membuat dataset dengan window size dan horizon
def create_dataset(data, window_size=5, horizon=1):
    X, y = [], []
    for i in range(len(data) - window_size - horizon + 1):
        X.append(data[i:(i + window_size), 0]) #0 adalah kolom Close
        y.append(data[(i + window_size + horizon - 1), 0]) #0 adalah kolom Close
    return np.array(X), np.array(y)

# Membuat dataset
window_size = 5
horizon = 1
X, y = create_dataset(dataAMD, window_size, horizon)
```

```python
X_apple, y_apple = create_dataset(dataApple, window_size, horizon)
```

Split data

```python
# Membagi dataset menjadi train, validation, dan test set (80% train, 10% val, 10% test)
train_size = int(len(X) * 0.8)
val_size = int(len(X) * 0.1)
test_size = len(X) - train_size - val_size

X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:train_size + val_size], y[train_size:train_size + val_size]
X_test, y_test = X[train_size + val_size:], y[train_size + val_size:]

# Reshape input menjadi [samples, time steps, features]
X_train = X_train.reshape((X_train.shape[0], window_size, 1))
X_val = X_val.reshape((X_val.shape[0], window_size, 1))
X_test = X_test.reshape((X_test.shape[0], window_size, 1))
```

```python
# Membagi dataset menjadi train, validation, dan test set (80% train, 10% val, 10% test)
train_sizeApple = int(len(X_apple) * 0.8)
val_sizeApple = int(len(X_apple) * 0.1)
test_sizeApple = len(X_apple) - train_sizeApple - val_sizeApple

X_trainApple, y_trainApple = X_apple[:train_sizeApple], y_apple[:train_sizeApple]
X_valApple, y_valApple = X_apple[train_sizeApple:train_sizeApple + val_sizeApple], y_apple[train_sizeApple:train_sizeApple + val_sizeApple]
X_testApple, y_testApple = X_apple[train_sizeApple + val_sizeApple:], y_apple[train_sizeApple + val_sizeApple:]

# Reshape input menjadi [samples, time steps, features]
X_trainApple = X_trainApple.reshape((X_trainApple.shape[0], window_size, 1))
X_valApple = X_valApple.reshape((X_valApple.shape[0], window_size, 1))
X_testApple = X_testApple.reshape((X_testApple.shape[0], window_size, 1))
```

## b

Buatlah arsitektur baseline dengan LSTM (units=50) dan layer akhir berupa
node Perceptron dengan units=1. Activation function untuk LSTM menggunakan ReLU

amd

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.optimizers import Adam

# Membuat model LSTM
modelamd = Sequential()
modelamd.add(LSTM(units=50, activation='relu', input_shape=(window_size, 1)))
modelamd.add(Dense(units=1))

# Kompilasi model
modelamd.compile(optimizer='adam', loss='mean_squared_error')

# Melatih model
history = modelamd.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_val, y_val))

# Evaluasi model
loss = modelamd.evaluate(X_test, y_test)
print(f'Test Loss: {loss}')
```

apple

```python
# Membuat model LSTM
modelapple = Sequential()
modelapple.add(LSTM(units=50, activation='relu', input_shape=(window_size, 1)))
modelapple.add(Dense(units=1))

# Kompilasi model
modelapple.compile(optimizer='adam', loss='mean_squared_error')

# Melatih model
history = modelapple.fit(X_trainApple, y_trainApple, epochs=50, batch_size=32, validation_data=(X_valApple, y_valApple))

# Evaluasi model
loss = modelapple.evaluate(X_testApple, y_testApple)
print(f'Test Loss: {loss}')
```

## c

modifikasi arsitektur pada nomor 1b untuk mendapatkan unjuk kerja yang optimal (kalian dapat menambahkan atau mengurangi arsitektur tersebut, atau mengganti hyperparameter, atau menggunakan tuning pada hyperparameter). Jelaskan alasan kalian untuk menggunakan pendekatan yang kalian pilih.

amd

```python
from keras.callbacks import EarlyStopping

# Membuat model LSTM yang dimodifikasi
modelamd1 = Sequential()
modelamd1.add(LSTM(units=200, return_sequences=True, activation='relu', input_shape=(window_size, 1)))
modelamd1.add(Dropout(0.1))
modelamd1.add(LSTM(units=100, activation='relu'))
modelamd1.add(Dropout(0.1))
modelamd1.add(Dense(units=1))

# Kompilasi model dengan learning rate yang lebih rendah
optimizer = Adam(learning_rate=0.001)
modelamd1.compile(optimizer=optimizer, loss='mean_squared_error')

# Menggunakan early stopping untuk mencegah overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Melatih model dengan batch size yang lebih kecil
history = modelamd1.fit(X_train, y_train, epochs=300, batch_size=32, validation_data=(X_val, y_val), callbacks=[early_stopping])

# Evaluasi model
loss = modelamd1.evaluate(X_test, y_test)
print(f'Test Loss: {loss}')
```

Perubahan yang dilakukan pada:
- Perubahan early stopping untuk mencegah overfitting dan menghemat waktu train.
- Penerapan dropout mencegah overfitting dengan mematikan sejumlah neuron secara acak selama train
- Penambahan Layer dan Units: Menambahkan lebih banyak unit di layer LSTM pertama dan menambahkan satu lagi layer LSTM dengan Dropout di antaranya membantu model untuk menangkap pola temporal yang lebih kompleks dan mengurangi overfitting.

apple

```python
from keras.callbacks import EarlyStopping

# Membuat model LSTM yang dimodifikasi
modelapple1 = Sequential()
modelapple1.add(LSTM(units=200, return_sequences=True, activation='relu', input_shape=(window_size, 1)))
modelapple1.add(Dropout(0.1))
modelapple1.add(LSTM(units=100, activation='relu'))
modelapple1.add(Dropout(0.1))
modelapple1.add(Dense(units=1))

# Kompilasi model dengan learning rate yang lebih rendah
optimizer = Adam(learning_rate=0.001)
modelapple1.compile(optimizer=optimizer, loss='mean_squared_error')

# Menggunakan early stopping untuk mencegah overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Melatih model dengan batch size yang lebih kecil
history = modelapple1.fit(X_trainApple, y_trainApple, epochs=250, batch_size=32, validation_data=(X_valApple, y_valApple), callbacks=[early_stopping])

# Evaluasi model
loss = modelapple1.evaluate(X_testApple, y_testApple)
print(f'Test Loss: {loss}')
```

Perubahan yang dilakukan pada:
- Perubahan early stopping untuk mencegah overfitting dan menghemat waktu train.
- Penerapan dropout mencegah overfitting dengan mematikan sejumlah neuron secara acak selama train
- Penambahan Layer dan Units: Menambahkan lebih banyak unit di layer LSTM pertama dan menambahkan satu lagi layer LSTM dengan Dropout di antaranya membantu model untuk menangkap pola temporal yang lebih kompleks dan mengurangi overfitting.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Mengembalikan skala data test ke bentuk aslinya
y_test_rescaled = y_test.reshape(-1, 1)

# Prediksi menggunakan data test
y_predamd = modelamd.predict(X_test)

# Menghitung RMSE
rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_predamd))
print(f'RMSE: {rmse}')

# Menghitung MAE
mae = mean_absolute_error(y_test_rescaled, y_predamd)
print(f'MAE: {mae}')

# Menghitung MAPE
mape = np.mean(np.abs((y_test_rescaled - y_predamd) / y_test_rescaled)) * 100
print(f'MAPE: {mape}')
```

Model base mendapatkan hasil yang cukup baik dapat terlihat dari metrik:
- RMSE: Kesalahan prediksi rata-rata dalam unit yang sama dengan harga saham. RMSE sebesar 0.9 menunjukkan bahwa rata-rata kesalahan prediksi sekitar 0.9 dolar.
- MAE: Menunjukkan rata-rata kesalahan absolut antara prediksi dan nilai aktual. MAE sebesar 0.53 menunjukkan bahwa dalam banyak kasus, prediksi model  rata-rata meleset sekitar 0.53 dolar dari nilai aktual
- MAPE: Informasi tentang kesalahan prediksi dalam bentuk persentase dari nilai aktual. MAPE sebesar 2.81% menunjukkan bahwa rata-rata kesalahan prediksi adalah sekitar 2.81% dari nilai aktual. Dalam konteks harga saham, dari berbagai sumber mengatakan bahwa MAPE di bawah 5% sering dianggap cukup baik karena menunjukkan bahwa model dapat memprediksi harga dengan tingkat kesalahan yang relatif rendah.

```python
plt.figure(figsize=(10, 4))
plt.plot(dfAMD.index[-len(y_test):], y_test_rescaled, color='blue', label='Actual Price')
plt.plot(dfAMD.index[-len(y_test):], y_predamd, color='red', label='Predicted Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('AMD Stock Price Prediction')
plt.legend()
plt.show()
```
